# 📊 Data Understanding & Patterns

```python
#What is the average house price (MEDV)?
avg_price = df["MEDV"].mean()
print("Average House Price:", avg_price)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["MEDV"], kde=True)
plt.title("Distribution of House Prices (MEDV)")
plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.show()
```

## Insight: Average House Price
#### The average house price is approximately 22.53.
#### From the distribution plot, most house prices are concentrated between 20 and 25.
#### The distribution is slightly right-skewed, indicating that while most houses have moderate prices, there are a few houses with significantly higher prices.
#### This suggests the presence of premium properties that increase the overall average.

```python
#What is the maximum and minimum house price?
max_price = df["MEDV"].max()
min_price = df["MEDV"].min()
print("Max Price:", max_price)
print("Min Price:", min_price)
```

## Insight: Maximum and Minimum House Prices
#### The maximum house price is 50, while the minimum price is 5.
#### This indicates a wide range in house prices, suggesting significant variation in property values across different areas.
#### This variation may be influenced by factors such as location, number of rooms, and socio-economic conditions.

```python
#What is the average number of rooms (RM)?
avg_rooms = df["RM"].mean()
print("Average Number of Rooms:", avg_rooms)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["RM"], kde=True)
plt.title("Distribution of Number of Rooms")
plt.xlabel("Number of Rooms")
plt.ylabel("Frequency")
plt.show()
```

## Insight: Average Number of Rooms
#### The average number of rooms is approximately 6.28.
#### From the distribution plot, most houses have between 5.5 and 6.5 rooms, indicating a fairly consistent housing structure across the dataset.
#### The distribution appears approximately normal, with only a few houses having significantly higher numbers of rooms.
#### This suggests that most properties are medium-sized, with limited variation in room count.

```python
#What is the average age of houses (AGE)?
avg_age = df["AGE"].mean()
print("Average House Age:", avg_age)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["AGE"], kde=True)
plt.title("Distribution of House Age (AGE)")
plt.xlabel("Proportion of Old Houses")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["PTRATIO"], kde=True)
plt.title("Distribution of Pupil-Teacher Ratio")
plt.xlabel("PTRATIO")
plt.ylabel("Frequency")
plt.show()
```

```python
#What is the relationship between number of rooms (RM) and house price (MEDV)?
correlation = df["RM"].corr(df["MEDV"])
print("Correlation between RM and MEDV:", correlation)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["RM"], y=df["MEDV"])
plt.title("Relationship between Number of Rooms and House Price")
plt.xlabel("Number of Rooms (RM)")
plt.ylabel("House Price (MEDV)")
plt.show()
```

## Insight: Relationship between Number of Rooms and House Price
#### The correlation between RM and MEDV is حوالي 0.69, which shows a strong positive relationship.
#### From the scatter plot, I can see that as the number of rooms increases, the house price also tends to increase.
#### Most of the points are going upward, which confirms this relationship.
#### However, the points are not perfectly aligned, which means there are other factors that also affect the house price.
#### Overall, the number of rooms is an important factor in determining house prices.

```python
#Which variable has the highest correlation with MEDV?
correlation_matrix = df.corr()
correlation_with_price = correlation_matrix["MEDV"].sort_values(ascending=False)
print(correlation_with_price)
```

```python
plt.figure(figsize=(10,6))
correlation_with_price.drop("MEDV").sort_values().plot(kind="barh")
plt.title("Correlation of Features with House Price (MEDV)")
plt.xlabel("Correlation Value")
plt.ylabel("Features")
plt.show()
```

## Insight: Correlation Analysis with MEDV
#### From the bar chart, I can see that RM has the highest positive correlation with house price, which means houses with more rooms tend to be more expensive.
#### On the other hand, LSTAT has the strongest negative correlation, which shows that areas with higher lower-status population tend to have lower house prices.
#### Other variables like PTRATIO, TAX, and INDUS also show negative relationships, but their effect is not as strong as LSTAT.
#### Overall, RM and LSTAT are the most important features affecting house prices in this dataset.

```python
#Which town has the highest median value of owner-occupied homes?
max_price = df["MEDV"].max()
town_index = df[df["MEDV"] == max_price]
print(town_index)
```

```python
top_prices = df[df["MEDV"] == df["MEDV"].max()]
plt.figure(figsize=(10,5))
plt.bar(top_prices.index, top_prices["MEDV"])
plt.title("Top Houses with Highest Price")
plt.xlabel("Town Index")
plt.ylabel("House Price (MEDV)")
plt.show()
```

## Insight: Towns with Highest House Price
#### From the analysis, I found that the highest median house price (MEDV) is 50.
#### There are multiple rows that have this value, which means more than one area has the same highest price.
#### This suggests that several towns share the top housing value.
#### Also, the value 50 appears many times, which might indicate that the dataset has a maximum cap for house prices.
#### So, instead of one specific town, there are multiple towns with the highest recorded house price.

```python
#How does LSTAT (lower-status population) affect house prices?
correlation = df["LSTAT"].corr(df["MEDV"])
print("Correlation between LSTAT and MEDV:", correlation)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["LSTAT"], y=df["MEDV"])
plt.title("Effect of LSTAT on House Price")
plt.xlabel("LSTAT (Lower Status Population)")
plt.ylabel("House Price (MEDV)")
plt.show()
```

## insight: Effect of LSTAT on House Price
#### From the scatter plot, I can clearly see a negative relationship between LSTAT and MEDV.
#### As LSTAT increases, the house price tends to decrease.
#### Most of the points are going downward, which confirms this relationship.
#### Although there is some variation in the data, the overall trend is very clear.
#### This shows that LSTAT has a strong impact on house prices in this dataset.

```python
#What is the relationship between RM (rooms) and LSTAT (lower-status population)?
correlation = df["RM"].corr(df["LSTAT"])
print("Correlation between RM and LSTAT:", correlation)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["RM"], y=df["LSTAT"])
plt.title("Relationship between Number of Rooms and LSTAT")
plt.xlabel("Number of Rooms (RM)")
plt.ylabel("LSTAT (Lower Status Population)")
plt.show()
```

## Insight: Relationship between RM and LSTAT
#### From the scatter plot, I can see a clear negative relationship between RM and LSTAT.
#### As the number of rooms increases, the value of LSTAT decreases.
#### This means that areas with larger houses tend to have a lower percentage of lower-status population.
#### The points are mostly going downward, which confirms this relationship.
#### So, RM and LSTAT are strongly related, and they both help explain differences in house prices.
